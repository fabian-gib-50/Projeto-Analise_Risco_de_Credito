   - A variável external_data_provider_email_seen_before possui um valor mínimo em -999, provavelmente este campo recebeu este valor, porque o sistema de cadastro não permitia valores nulos para preenchimento. Portanto, estes valores serão considerados como NaN;

   - A variável credit_limit possui um mínimo de 0 reais, não é uma prática do mercado de cartão de crédito liberar apenas 0 reais de limite. Portanto, este valor será substituído por NaN;
   - Criaremos uma variável de agrupamento por região do Brasil, utilizando a variável shipping_state;
   - A variável ids serve para identificar um determinado cliente, portanto, não agrega valor no objetivo proposto e esta coluna será excluída;
   - A variável channel apenas tem um valor único, portanto, não agrega valor a análise e será excluída;
   - As variáveis lat_lon e shipping_zip_code serão excluídas do dataset para simplificar a modelagem.
     - **`O próximo passo é excluir os valores ausentes da variável alvo target_default e as colunas mencionadas na análise anterior. Além disso, vamos substituir por NaN, os valores negativos, infinitos e 0 das variáveis external_data_provider_email_seen_before, reported_income e credit_limit, respectivamente`**.

```python
# Optei por remover a variável "credit_limit", para esse projeto não precisarei dela

def preencher_proporcional( cols ): 
    
    """ 
    @DbianDs
    Preenche valores ausentes na mesma proporção dos valores presentes
    Recebe uma coluna e retorna a coluna com os valores ausentes preenchidos
    na proporção dos valores previamente existentes 
                                                   """
    
# Gerando um dicionário com valores únicos e sua porcentagens
    
    percentages = cols.value_counts( normalize=True ).to_dict()

# Tranformando as chaves e valores do dicionário em listas 
    
    percent = [ percentages[key] for key in percentages ]
    labels  = [ key for key in percentages ]

# Utilizando as listas para prencher os valores nulos na proporção correta 
    
    series = pd.Series( np.random.choice( labels, p=percent, size=col.isnull().sum()) )
    cols = col.fillna( series )
    
# Verificando se todos os valores ausentes foram preenchidos e
# preenchendo os que não tiverem sido

    if len( cols.isnull() ) > 0:
        cols.fillna( value=max( percentages, key=percentages.get ), inplace=True, axis=0 )
    return cols

# Preenchendo os valores ausentes
 
    """ 
       Para a coluna credito_clean, 
       localize pelo índice 1, liste coluna  
       se credito_clean, coluna com tipo == 0 
       preencer proporcional coluna credito_clean
       senão aplicar metodo fillna com valores de média
       no lugar com método inplace na linha com eixo 0
                                                   """
    for col in credito_clean.iloc[ :,1: ].columns.tolist():
            if  credito_clean[ cols ].dtypes == '0':
                credito_clean[ cols ] = preencher_proporcional( credito_clean[ cols ] )
            else:
                credito_clean[ cols ].fillna( value=credito_clean[ cols ].median(), inplace=True, axis=0)
```

  - Simétrica: Uma distribuição pode ser chamada de simétrica se parecer a mesma à esquerda e à direita do ponto central.
  - Assimétrica: Uma distribuição pode ser chamada de assimétrica se não parecer a mesma à esquerda e à direita do ponto central.
    
  - Curtose:
    - É também um termo estatístico e uma característica importante da distribuição de frequência. Determina se uma distribuição é de cauda pesada em relação à distribuição normal. Ele fornece informações sobre a forma de uma distribuição de frequência.

    - curtose para distribuição normal é igual a 3.
  - Para uma distribuição com curtose < 3: é chamada playkurtic.
  - Para uma distribuição com curtose > 3, é chamada de leptocúrtica e significa que ela tenta produzir mais valores discrepantes do que a distribuição normal.

  - Boxplot são gráficos que nos mostam a distribuição dos valores numéricos, podendo ser `int ou float`, bem como autliers ( valores acima da média ) para que possamos tratá-los devidamente.

```python
# For para implemntar "boxplots"

var_num = [ ]

for i in df_clean.columns[ 0:16 ].tolist():
           if df_clean.dtypes[ i ] == 'int64' or df_clean.dtypes[i] == 'float64':
# print( i, ':', df_raw.dtypes[ i ] )
              var_num.append( i )
```

### Considerações sobre os BoxPlots: 

   - Com os Boxplots é possível ver que temos a presença de `outliers` em algumas features. Porém, não aplicarei o método ( **IQR** ) score para retirar estas anomalias como por exemplo as variáveis score, risk_rate entre outras, que podem inviabilizar o modelo de `Machine Mearning`. Aplicar o método ( **IQR** ) apenas nas seguintes variáveis: income, reported_income, n_accounts e n_issue.

```python
# Irei armazenar as features para aplicar o método ( features IQR )

features_IQR = [ 
               'income',
               'reported_income',
               'n_accounts',
               'n_issues'
               ] 

# Criando looping nas features selecionadas para o for 

for feature in features_IQR: 
    
  Q1  = df_clean[feature].quantile( q = 0.25 ) # Definindo o primeiro quartil
  Q3  = df_clean[feature].quantile( q = 0.75 ) # Definindo o segundo quartil
  IQR = Q3 - Q1                                # Definindo o interquartil
  Limite_Superior = Q3 + IQR                   # Definindo o Limite Superior
  Limite_Inferior = Q1 - IQR                   # Definindo o Limite Inferior

# Efeturando limpeza dos outliers ácima do limite superior

df_clean.drop( df_clean.loc[ df_clean[feature] > Limite_Superior ].index, axis = 0, inplace = True )

# Efeturando limpeza dos outliers abaixo do limite superior 

df_clean.drop( df_clean.loc[ df_clean[feature] < Limite_Inferior ].index, axis = 0, inplace = True )
```

### Aplicando o teste de Tukey nas variáveis numéricas

___
  
  - Após a limpeza e ajustes das variáveis numéricas, aplicarei o método `Teste de Tukey`. 
  - O método Tukey, aplicado em análise estatística, é importante para a observância da distribuição das variáveis numéricas, bem como seus limites, com aplicação de parâmetros específicos.
  - Dentre os testes de comparações múltiplas mais utilizados, o Teste de Tukey se destaca por ser poderoso ao fazer comparações entre todos os pares e também por ser de fácil aplicação. Também é conhecido como Teste de Tukey HSD (Teste de Tukey da Diferença Honestamente Significativa).

   - O teste de Tukey foi desenvolvido por John Wilder Tukey e apresentado em 1949 no artigo titulado “Comparing Individual Means in the Analysis of Variance” (Biometrics. 5 (2): 99–114. JSTOR 3001913).
  
   - Para realizar o Teste de Tukey, deve ser levada em conta as seguintes suposições:

     - As observações são independentes dentro e entre os grupos;
     - Os grupos devem ser normalmente distribuídos;
     - A variância dentro do grupo deve ser constante.
     - Farei alguns agrupamentos e também algumas análises com foco em crédito, renda e etc. O objetivo é extrair o máximo de insights possíveis dessa base.

   - Exemplo: todos os celulares que acessam a base via app ou web, deixam log de registros, logo, posso com esses dados extrair informações como por exemplo, a quantidade de clientes que acessaram os canais, e pelo sistema operacional consigo determinar o modelo de celular, podendo gerar insights de qualificação para os times de marketing digital, desencadear campanhas e etc...

```python
# Função para mapear o sistema operacional dos clientes, pela variável ( user_agent ), trata-se de uma ( String )

def mapeamento_sist_op ( string ):

# Map list dos sistemas operacionais

  List_so = [ 'android',
              'iphone',
              'ubuntu',
              'windows',
              'macintosh' 
            ]

# Criarei uma variável para armazenar os resultados obtivos com def ( list_so ) 

  result = ""

# Passando a string para minúsculo 

  string_lower = string.lower()

# Gero com Python um loop em relação a lista 

  for so in List_so:

# Verifica se o sistema está contido no conjunto de caracteres 

    if so in string_lower:

# Armazena os dados do map so sistema operacinal dos clientes
      result = so
  
# A funcion Python retorna o resultado do mapeamento da var "user_agent"

  if result == "":
    return "Outro"
  else:
    return result

# Eng. atributos: 
# crio a variável que irá receber o looping Pyhon para visualização na base, passando para ela "mapeamento_sist_op"

df_clean['SO'] = df_clean[ 'user_agent' ].apply( mapeamento_sist_op )


# Eliminando a variável origem para não gerar redundância de variáveis e consequentemente comprometer os modelos de I.A
# Com o método inplace, efeturo o ajuste na raiz, base.

df_clean.drop( labels = [ 'user_agent' ], axis = 1, inplace = True )
     
```

  - Esta etapa pode ser feita por duas maneiras, com **Over-sampling**: cria novas observações da classe minoritária a partir das informações contidas nos dados originais. Essa geração de novas entradas pode ser feita aleatoriamente com o auxílio de técnicas de clustering ou sinteticamente.
**Under-sampling**: reduz o desbalanceamento do dataset focando na classe majoritária. Ou seja, elimina aleatoriamente entradas da classe com maior número de ocorrências.
Utilizaremos como base a técnica de **under-sampling**, devido a possibilidade de mantermos integralmente os dados originais associados a **target_default**.

```python
# Aplicação dos dados de treino e testes, com ( "train_X e test_y", predicao e variavel_target )

X = df_eng.drop( [ 'target_default' ], axis = 1 ).select_dtypes( exclude= 'object' )
y = df_eng[ 'target_default' ] 

train_X, test_X, train_y, test_y = train_test_split( X, y, shuffle = True, 
                                                             stratify = y, 
                                                          test_size = 0.3, 
                                                       random_state = 40 )

# Chegando as dimensões de ( X e y ), para treino e testes 

print( 'Dimensões de ( X e y ), em dados de treino e testes\n' )
print( train_X.shape , train_y.shape, test_X.shape, test_y.shape )
```

### Modelo entrando em produção com dados de treino e teste com I.A

   - Métricas e Modelos

Para decidirmos que métrica usar, precisamos analisar algumas de nossas opções.

* Caso optemos pela `Acurácia`, precisamos estar cientes que esse pode não ser o método ideal, mas para uma análise inicial do problema, pode nos ajudar a criar uma solução generalista.

* Caso optemos por otimizar o modelo a partir do `Recall`, estamos priorizando que não haja Falsos Negativos. Ou seja, é preferível negar um cliente que não traria prejuízo à empresa, do que aceitar um que traga. Essa é a forma que são tratados, por exemplo, fraudes em cartão de crédito, modelos médicos, etc.

* Caso optemos por otimizar a partir do `Precision`, essencialmente, estamos dizendo que Falsos Positivos são ok, e não queremos ter Falsos Negativos. Basicamente, é preferível aceitar clientes que possam trazer prejuízo, do que deixar de aceitar clientes que não trarão. Essa é a forma que são tratados, por exemplo, modelos para spam de e-mails.

Aqui, vamos utilizar a `Acurácia`, no intuito de realizar uma análise generalista, deixando o *fine-tuning* de métricas e modelos para o talento dos alunos.

O que vai definir a melhor métrica é a estratégia de negócio da empresa, e seus valores. Também por isso, vamos utilizar essa métrica mais "genérica".

Dito isso, mãos à obra!

```python
# Construindo a função de avaliação do baseline, para enriquecer o modelo e ajustar o pipeline dos modelos, passando para ( X e y ) o clf ( cross validation )

def valididacao_model ( X, y, clf, quite=False ):
    """
    @Fabiands
    Realiza cross-validation com os dados de treino para determinado modelo.

    # Arguments
        X: DataFrame, contém as variáveis independetes.
        y: Series, vetor contendo a variável alvo.
        clf: modelo classificador do Scikit-learn.
        quite: bool, indicando se a função deve imprimir os resultados ou não.

    # Returns
        float, média dos scores da cross-validation.
    """
    
    X = np.array( X )
    y = np.array( y )

    pipeline = make_pipeline( StandardScaler(), clf )
    scores = cross_val_score( pipeline, X, y, scoring='accuracy' )

    if quite == False:
        print( 'Acurácia: {:.4f} ( +/- {:.2f} )'.format( scores.mean(), scores.std()) )
    
    return scores.mean()

# Criação do baseline e verificação do desempenho para implementações do modelo de predição com I.A, com Random Forest ( rf )

print( '' )
print( 'Resultado Random Forest Classifier:' )
rf = RandomForestClassifier()
score_baseline = valididacao_model( train_X, train_y, rf )
```
